### Medium complexity - Birthday Paradox

Birthday paradox concerns the probability that, in a set of `n` randomly chosen people, some pair of them will have the same birthday. You can read about this problem at https://en.wikipedia.org/wiki/Birthday_problem.

Write a function called **`has_duplicates()`**, that takes a list and returns `True` if there is any element that appears more than once. It should not modify the original list.

If there are 45 students in the class, what are the chances that any two students will have the same birthday? You can estimate this probability by generating random samples of 45 birthdays and checking for matches.

**Hints:**
1. you can generate random birthdays with the `randint` function in the `random` module. 
2. For simplicity, use the day number of the year, not the actual date
3. you can use the book solution as a starting point for this assignment: https://github.com/AllenDowney/ThinkPython2/blob/master/code/birthday.py or http://greenteapress.com/thinkpython2/code/birthday.py

The code should print out: 
- number of students
- number of iterations/samples
- list of duplicate days for each iteration, where duplicates are found
- probability

```python
""" Below function called has_duplicates(), takes a list and returns True
if there is any element that appears more than once without modifying the original list"""
def has_duplicates(t):
     #t1 is a copy of t to avoid modifying the origional list
    t1 = t[:]
    #list is sorted first because it is easier to do a comaprison when list is in order
    t1.sort()

    # check for adjacent elements that are equal
    for i in range(len(t1)-1):
        if t1[i] == t1[i+1]:
            return True
    return False

"""Below function returns a list of random birth dates for given number of students """
from random import randint
def random_bdays(num_students):
    # bdays a new list is defined to store list of birthdays generated by randint function
    bdays=[]
    for i in range(num_students):
        bdays.append(randint(1,365))
    return bdays

"""Below function takes a list as a parameter and retun a list of duplicate elements in that list"""
def duplicate_values(d):
    #t is a copy of d to avoid modifying the parameter
    t=d[:]
    #dupes is an empty list defined to collect duplicate elements using append method from list t
    dupes=[]
    t.sort()
    for i in range(len(t)-1):
        if t[i]==t[i+1]:
            dupes.append(t[i])
    return dupes

"""Below function takes an integer number of students and returns probablity of having two students the same birthday,
probability formula has been implemented from - https://en.wikipedia.org/wiki/Birthday_problem """
def same_birthday_probability(n):
    #p is initialized to 1 so that the below formula can work other wise python will throw an error p is used before assigned
    p=1
    for i in range(n):
        p=p*((365-i)/365)
    pf=1-p
    #below is rounding the probabilty into percentage and upto 4 decimal places
    pf=round(100*pf,4)
    return pf


"""Below is the main function that executes a For loop to traverse through different iterations/samples generated by random_bday
function.
It makes a function call has_duplicates to check if within each list (set of generated birthdays) has any duplicate days and
if so it will return true.
If has_duplicates function returns true then it appends the list of elements to the empty list dupes by making a function call
duplicate_values, which passes the list of current iteration and returns a list of duplicate elements from that list
"""

def main():
    num_of_students = 45
    num_of_iterations = 10
    #optional count is initialized to find in how many iterations there are duplicates
    count=0
    dupes=[]
    for i in range(num_of_iterations):
        bdays=random_bdays(num_of_students)
        #below print statement prints the list of all samples with the iteration numbers
        print('Iteration sample %d \n' %(i+1),bdays)
        if has_duplicates(bdays) == True:
        #below print statement can be un commented to verify the duplciate elements per iteration as it will indicate  
        #the iteration which contains duplicate elements.
            #print ('The above iteration %d has duplicates\n'%(i+1))
            dupes.append(duplicate_values(bdays))
            count=count+1
    print('The number of students in a class are:', num_of_students)
    print('The number of iterations:', num_of_iterations)
    print('There are at least %d iterations with duplicate elements'%count)
    print('The list of duplicates includes duplicates for all iterations and a sub list of duplicates for per iteration:\n',
          dupes)
    #same_birthday_probability function has been called within the print statement to find the probability of 2 students having
    #the same birthday in the class of 45 students
    print('The probability of having two students the same birthdate in the class of %d is'%num_of_students,
          same_birthday_probability(num_of_students),'%')

main()
```
